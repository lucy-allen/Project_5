```python
states = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'Florida', 
         'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 
         'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 
         'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio', 
         'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 
         'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming']
```

```python
import numpy as np
```

```python
def state_function(location):
    state_name = ''
    for state in states:
        if state in location:
            state_name = state
        elif 'Lake Tahoe' in location:
            state_name = 'California'
    if state_name != '':
        return state_name
    else:
        return np.nan
```

```python
def Elevations(info):
    new_info = info.split('    ')
    return new_info[2]
```

```python
def Elevations2(info):
    new_info = info.split('    ')
    if len(new_info) >= 5:
        return new_info[2]
    else:
        return np.nan
```

```python
resort_data['Elevation_Info'] = resort_data['Info'].map(Elevations)
```

```python
resort_data2['Elevation_Info'] = resort_data2['Info'].map(Elevations2)
```

```python
def Distance(info):
    new_info = info.split('    ')
    return new_info[3]
```

```python
def Distance2(info):
    new_info = info.split('    ')
    if len(new_info) >= 5:
        return new_info[3]
    else:
        return np.nan
```

```python
resort_data['Distance_Info'] = resort_data['Info'].map(Distance)
```

```python
resort_data2['Distance_Info'] = resort_data2['Info'].map(Distance2)
```

```python
def Lifts(info):
    new_info = info.split('    ')
    return new_info[4]
```

```python
def Lifts2(info):
    new_info = info.split('    ')
    if len(new_info) >= 5:
        return new_info[4]
    else:
        return np.nan
```

```python
resort_data['Lift_Info'] = resort_data['Info'].map(Lifts)
```

```python
resort_data2['Lift_Info'] = resort_data2['Info'].map(Lifts2)
```

```python
def Prices(info):
    new_info = info.split('    ')
    return new_info[5]
```

```python
def Prices2(info):
    new_info = info.split('    ')
    if len(new_info) >= 6:
        return new_info[5]
    else:
        return np.nan
```

```python
resort_data['Price_Info'] = resort_data['Info'].map(Prices)
```

```python
resort_data2['Price_Info'] = resort_data2['Info'].map(Prices2)
```
