### Elevation Functions

```python
### Vertical Distance
def vertical_distance(info):
    info_list = info.split('m')
    vert = info_list[0]
    return int(vert)
```

```python
def base_elevation(info):
    info_list = info.split('m')
    base = info_list[1]
    base_elev = ''
    for char in base:
        if char.isnumeric():
            base_elev += char
    return int(base_elev)
```

```python
def peak_elevation(info):
    info_list = info.split('m')
    peak = info_list[2]
    peak_elev = ''
    for char in peak:
        if char.isnumeric():
            peak_elev += char
    return int(peak_elev)
```

### Distance Functions

```python
### Total Distance
def Total_distance(info):
    info_list = info.split('km')
    total = info_list[0]
    return float(total)
```

```python
### Easy Distance
def Easy_distance(info):
    info_list = info.split('km')
    total = info_list[1]
    return float(total)
```

```python
### Intermediate Distance
def Intermediate_distance(info):
    info_list = info.split('km')
    total = info_list[2]
    return float(total)
```

```python
### Hard Distance
def Hard_distance(info):
    info_list = info.split('km')
    total = info_list[3]
    return float(total)
```

### Lift Function

```python
def lift_num(info):
    num = ''
    for char in info:
        if char.isnumeric():
            num += char
    return int(num)
```

### Price Function

```python
def price(info):
     if type(info) == type(5.4):
        return info   
     else:
        info_list = info.split(',')
        price = info_list[0]
        total_price = ''
        for char in price:
            if char.isnumeric():
                total_price += char
        return int(total_price)
```

##  combine all functions

```python
all_resorts['Vertical_Distance'] = all_resorts['Elevation_Info'].map(vertical_distance)
```

```python
all_resorts['Base_Elevation'] = all_resorts['Elevation_Info'].map(base_elevation)
```

```python
all_resorts['Peak_Elevation'] = all_resorts['Elevation_Info'].map(peak_elevation)
```

```python
all_resorts['Total_Distance'] = all_resorts['Distance_Info'].map(Total_distance)
```

```python
all_resorts['Easy_Distance'] = all_resorts['Distance_Info'].map(Easy_distance)
```

```python
all_resorts['Intermediate_Distance'] = all_resorts['Distance_Info'].map(Intermediate_distance)
```

```python
all_resorts['Hard_Distance'] = all_resorts['Distance_Info'].map(Hard_distance)
```

```python
all_resorts['Lifts'] = all_resorts['Lift_Info'].map(lift_num)
```

```python
all_resorts['Price'] = all_resorts['Price_Info'].map(price)
```

```python
all_resorts.drop(['Elevation_Info', 'Distance_Info', 'Lift_Info', 'Price_Info'], axis=1, inplace=True)
```

### Percent Values for Distances

```python
def percentages(num, denom):
    return (num/denom)*100
```

```python
all_resorts['Pct_Easy'] = percentages(all_resorts['Easy_Distance'], all_resorts['Total_Distance'])    
```

```python
all_resorts['Pct_Intermediate'] = percentages(all_resorts['Intermediate_Distance'], all_resorts['Total_Distance'])    
```

```python
all_resorts['Pct_Hard'] = percentages(all_resorts['Hard_Distance'], all_resorts['Total_Distance'])    
```
